```python
import urllib.request
import pandas as pd
import json

# 클라이언트 ID와 시크릿 설정
client_id = ""
client_secret = ""

# API URL
url = "https://openapi.naver.com/v1/datalab/shopping/categories"

# 요청할 데이터 설정
request_data_template = {
    "startDate": "2024-01-01",
    "endDate": "2024-10-20",
    "timeUnit": "date",
    "category": [
        {"name": "원피스", "param": ["50000807"]},
        {"name": "상의", "param": ["50000803"]},
        {"name": "하의", "param": ["50000808"]}

    ],
    "device": "",  # 빈 문자열로 두어 모든 기기
    "gender": "",  # 빈 문자열로 두어 모든 성별
    "ages": []  # 빈 리스트로 두어 특정 연령대
}

# 요청 헤더 설정
headers = {
    "X-Naver-Client-Id": client_id,
    "X-Naver-Client-Secret": client_secret,
    "Content-Type": "application/json"
}

# 데이터 저장 리스트
rows = []

# 기기, 성별, 연령대 설정
devices = ['pc', 'mo']
genders = ['f', 'm']
ages = [['10'], ['20'], ['30'], ['40'], ['50'], ['60']]  # 다양한 연령대

for device in devices:
    for gender in genders:
        for age_group in ages:
            request_data = request_data_template.copy()  # 템플릿 복사
            request_data['device'] = device
            request_data['gender'] = gender  # 성별 설정
            request_data['ages'] = age_group  # 각 연령대 설정

            # JSON 데이터로 인코딩
            data = json.dumps(request_data).encode('utf-8')

            # API 요청
            request = urllib.request.Request(url, data=data, headers=headers)

            try:
                response = urllib.request.urlopen(request)
                rescode = response.getcode()

                if rescode == 200:
                    response_body = response.read()
                    response_dict = json.loads(response_body.decode('utf-8'))

                    # 응답 데이터 출력
                    # print(json.dumps(response_dict, indent=4, ensure_ascii=False))  # 응답 내용 출력 (디버깅용)

                    # 응답 데이터 처리
                    results = response_dict.get('results', [])
                    if not results:
                        print("No results found.")
                        continue

                    for category in results:
                        category_name = category.get('title', 'N/A')  # 카테고리 이름 추출
                        category_param = request_data['category'][0]['param']  # 카테고리 코드 추출
                        for item in category.get('data', []):
                            rows.append({
                                'StartDate': request_data['startDate'],
                                'EndDate': request_data['endDate'],
                                'TimeUnit': request_data['timeUnit'],
                                'Category': category_name,
                                'CategoryParam': ', '.join(category_param),  # 카테고리 코드 추가
                                'Period': item['period'],
                                'Ratio': item['ratio'],
                                'Device': device,
                                'Gender': gender,
                                'Ages': ', '.join(request_data['ages'])
                            })

                else:
                    print(f"Error Code {rescode}: {response.read().decode('utf-8')}")  # 오류 메시지 출력
                    break

            except Exception as e:
                print(f"Error: {e}")
                break

# 모든 데이터를 DataFrame으로 변환
df = pd.DataFrame(rows)

# 총 데이터 개수 출력
print(f"총 데이터 개수: {len(df)}")  # DataFrame의 길이 출력

# DataFrame 출력
print(df.to_string(index=False))  # 인덱스 없이 깔끔하게 출력

# DataFrame을 CSV 파일로 저장
csv_file_name = "쇼핑인사이트 분야별 트렌드 조회(상의,하의,원피스).csv"
df.to_csv(csv_file_name, index=False, encoding='utf-8-sig')  # UTF-8 인코딩으로 저장
print(f"데이터가 '{csv_file_name}' 파일로 저장되었습니다.")
```
